```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df_normal = pd.read_csv('eval_gpt_4o_mini_persian_qwen1.7b.csv')  # Replace with your file
df_masked = pd.read_csv('eval_gpt_4o_mini_persian_masked_qwen1.7b.csv')  # Replace with your file
```

```python
df1 = df_normal.copy()
df2 = df_masked.copy()
```

```python
comparison_df = pd.DataFrame({
    "is_correct_exp_1": df1["is_correct_exp"],
    "is_correct_exp_2": df2["is_correct_exp"],
    "is_correct_noexp_1": df1["is_correct_noexp"],
    "is_correct_noexp_2": df2["is_correct_noexp"],
    "sufficiency_1": df1["sufficiency"],
    "sufficiency_2": df2["sufficiency"],
    "ppl_qwen_1": df1["ppl_qwen"],
    "ppl_qwen_2": df2["ppl_qwen"]
})
```

```python
summary_stats = {
    "Accuracy_Exp": [df1["is_correct_exp"].mean(), df2["is_correct_exp"].mean()],
    "Accuracy_NoExp": [df1["is_correct_noexp"].mean(), df2["is_correct_noexp"].mean()],
    "Sufficiency": [df1["sufficiency"].mean(), df2["sufficiency"].mean()],
    "Avg_PPL": [df1["ppl_qwen"].mean(), df2["ppl_qwen"].mean()]
}
summary_df = pd.DataFrame(summary_stats, index=["Model 1", "Model 2"])
summary_df
```

```python
melted = pd.melt(comparison_df[["ppl_qwen_1", "ppl_qwen_2"]], var_name="Model", value_name="Perplexity")

sns.histplot(data=melted, x="Perplexity", hue="Model", bins=30, kde=True)
plt.title("Perplexity Distribution: Qwen")
plt.xlabel("Perplexity")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
acc_df = pd.DataFrame({
    "Metric": ["Accuracy (Exp)", "Accuracy (NoExp)", "Sufficiency"],
    "Model 1": [
        df1["is_correct_exp"].mean(),
        df1["is_correct_noexp"].mean(),
        df1["sufficiency"].mean()
    ],
    "Model 2": [
        df2["is_correct_exp"].mean(),
        df2["is_correct_noexp"].mean(),
        df2["sufficiency"].mean()
    ]
})

acc_df.set_index("Metric").T.plot(kind="bar")
plt.title("Performance Comparison")
plt.ylabel("Rate")
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
