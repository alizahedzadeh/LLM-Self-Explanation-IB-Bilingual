```python
import re

def extract_prediction(raw_output: str):
    if not isinstance(raw_output, str):
        return None
    
    # Try <prediction>X</prediction>
    match = re.search(r"<prediction>\s*([A-D])\s*</prediction>", raw_output, re.IGNORECASE)
    if match:
        return match.group(1).upper()

    # Try [A], (A), {A}, "A", etc.
    match = re.search(r"[\(\[\{]?\s*([A-D])\s*[\)\]\}]?", raw_output, re.IGNORECASE)
    if match:
        return match.group(1).upper()
    
    # Try "Answer: A" or "Prediction: B"
    match = re.search(r"(answer|prediction)\s*[:\-]?\s*([A-D])", raw_output, re.IGNORECASE)
    if match:
        return match.group(2).upper()
    
    return None
```

```python
grok_en['prediction'] = grok_en['raw_output'].apply(extract_prediction)
```
