```python
import pandas as pd
import random
from IPython.display import display, HTML

# Function to display each result with LTR formatting and dynamic choices
def display_result_ltr(row):
    # Dynamically format the choices based on the number of options available
    choices_html = "".join([f"<br>{i+1}: {choice}" for i, choice in enumerate(row['choices'])])

    html = f"""
    <div style='border: 2px solid black; padding: 10px; margin-bottom: 10px; direction: ltr; text-align: left;'>
        <p><strong>Question:</strong> {row['question']}</p>
        <p><strong>Lecture:</strong> {row['lecture']}</p>
        <p><strong>Topic:</strong> {row['topic']}</p>
        <p><strong>Category:</strong> {row['category']}</p>
        <p><strong>Skill:</strong> {row['skill']}</p>
        <p><strong>Grade:</strong> {row['grade']}</p>
        <p><strong>Options:</strong> {choices_html}</p>
        <p><strong>Correct Answer:</strong> {row['answer']}</p>
        <p><strong>Solution:</strong> {row['solution']}</p>
    </div>
    """
    display(HTML(html))

# Function to sample and display questions from each topic
def sample_questions_by_topic(df, num_samples=3):
    """
    Samples a few questions from each topic in the dataset.
    """
    grouped = df.groupby("topic")
    for topic, group in grouped:
        print(f"\nüîπ **Topic: {topic}** ({len(group)} questions)\n")
        sampled_rows = group.sample(min(num_samples, len(group)))  # Sample questions
        for _, row in sampled_rows.iterrows():
            display_result_ltr(row)

# Example usage with the ScienceQA dataset (Assuming `df` is the cleaned dataset)
sample_questions_by_topic(df_text_only, num_samples=2)  # Adjust number of samples if needed
```

# Some Feature Engineering 

```python
import numpy as np

# 1Ô∏è‚É£ ÿ™⁄©ŸÖ€åŸÑ ŸÖŸÇÿßÿØ€åÿ± ÿÆÿßŸÑ€å ÿØÿ± choice_3 Ÿà choice_4
df["choice_3"].fillna("None", inplace=True)
df["choice_4"].fillna("None", inplace=True)

# 2Ô∏è‚É£ Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß€å ÿ≤ÿ®ÿßŸÜ€å ÿ≥ÿ§ÿßŸÑ
df["question_length"] = df["question"].apply(len)
df["num_words"] = df["question"].apply(lambda x: len(x.split()))
df["num_sentences"] = df["question"].apply(lambda x: x.count(".") + x.count("?") + x.count("!"))
df["avg_word_length"] = df["question"].apply(lambda x: np.mean([len(word) for word in x.split()]))

# 3Ô∏è‚É£ ÿ™ÿ¥ÿÆ€åÿµ ŸÜŸàÿπ ÿ≥ÿ§ÿßŸÑ
def categorize_question_type(q):
    q_lower = q.lower()
    if "what is" in q_lower or "define" in q_lower:
        return "definition"
    elif "why" in q_lower or "cause" in q_lower:
        return "causal"
    elif "how is" in q_lower or "compare" in q_lower:
        return "comparison"
    elif "is" in q_lower or "does" in q_lower:
        return "yes/no"
    elif "how much" in q_lower or "calculate" in q_lower:
        return "calculation"
    else:
        return "other"

df["question_type"] = df["question"].apply(categorize_question_type)

# 4Ô∏è‚É£ ÿ™ÿπ€å€åŸÜ ÿ≥ÿ∑ÿ≠ ÿ≥ÿÆÿ™€å ÿ®ÿ± ÿßÿ≥ÿßÿ≥ ÿ∑ŸàŸÑ Ÿà Ÿæ€å⁄Ü€åÿØ⁄Ø€å ÿ≥ÿ§ÿßŸÑ
def determine_difficulty(length, sentences):
    if length < 50 and sentences == 1:
        return "easy"
    elif length < 100 and sentences <= 2:
        return "medium"
    else:
        return "hard"

df["difficulty_level"] = df.apply(lambda row: determine_difficulty(row["question_length"], row["num_sentences"]), axis=1)

# 5Ô∏è‚É£ Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß€å ŸÖÿ±ÿ®Ÿàÿ∑ ÿ®Ÿá Ÿæÿßÿ≥ÿÆ ÿµÿ≠€åÿ≠
df["answer_length"] = df["original_answer"].apply(len)
df["answer_is_longest"] = df.apply(lambda row: 1 if len(row["original_answer"]) == max(len(row["choice_1"]), len(row["choice_2"]), len(row["choice_3"]), len(row["choice_4"])) else 0, axis=1)
df["answer_is_shortest"] = df.apply(lambda row: 1 if len(row["original_answer"]) == min(len(row["choice_1"]), len(row["choice_2"]), len(row["choice_3"]), len(row["choice_4"])) else 0, axis=1)

# 6Ô∏è‚É£ €åÿßŸÅÿ™ŸÜ ÿ≥ÿ§ÿßŸÑÿßÿ™ ŸÖÿ¥ÿßÿ®Ÿá ÿ®ÿß NLP
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# ÿ™ÿ®ÿØ€åŸÑ ÿ≥ŸàÿßŸÑÿßÿ™ ÿ®Ÿá ÿ®ÿ±ÿØÿßÿ±Ÿáÿß€å ÿπÿØÿØ€å (Embedding)
model = SentenceTransformer('all-MiniLM-L6-v2')
question_embeddings = model.encode(df["question"].tolist())

# ŸÖÿ≠ÿßÿ≥ÿ®Ÿá ÿ¥ÿ®ÿßŸáÿ™ ⁄©ÿ≥€åŸÜŸàÿ≥€å ÿ®€åŸÜ ÿ≥ÿ§ÿßŸÑÿßÿ™
similarity_matrix = cosine_similarity(question_embeddings)
np.fill_diagonal(similarity_matrix, 0)  # ÿ≠ÿ∞ŸÅ ÿÆŸàÿØ ÿ¥ÿ®ÿßŸáÿ™€å

# €åÿßŸÅÿ™ŸÜ ÿ®€åÿ¥ÿ™ÿ±€åŸÜ ÿ¥ÿ®ÿßŸáÿ™ ÿ®ÿ±ÿß€å Ÿáÿ± ÿ≥ÿ§ÿßŸÑ
df["similar_question_id"] = np.argmax(similarity_matrix, axis=1) + 1  # ⁄ÜŸàŸÜ question_id ÿßÿ≤ 1 ÿ¥ÿ±Ÿàÿπ ÿ¥ÿØŸá

# ŸÜŸÖÿß€åÿ¥ ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿ®Ÿá‚Äåÿ±Ÿàÿ≤ÿ±ÿ≥ÿßŸÜ€å‚Äåÿ¥ÿØŸá
import ace_tools as tools
tools.display_dataframe_to_user(name="Enhanced Dataset with New Features", dataframe=df)
```
