```python
# update_sufficiency_usefulness.py
import pandas as pd
from pathlib import Path

EVAL_DIR   = Path("/home/a_zahedzadeh/self-explaination-thesis/trim_gpt_4o_mini/en/evaluation")      # پوشهٔ فایل‌ها
OVERWRITE  = True                    # اگر False باشد فایل جدید با _new ذخیره می‌کنیم

def update_file(csv_path: Path):
    df = pd.read_csv(csv_path)

    # ستون‌های مفقود را بساز
    if 'is_correct_noexp' not in df.columns:
        df['is_correct_noexp'] = df['pred_noexp'] == df['gold']
    if 'is_correct_exp' not in df.columns:
        df['is_correct_exp']   = df['pred_exp']  == df['gold']

    # تعریف‌های نهایی
    df['sufficiency'] = df['is_correct_exp']

    df['usefulness'] = (
        (~df['is_correct_noexp'] &  df['is_correct_exp']) |
        ( df['is_correct_noexp'] &  df['is_correct_exp'] &
          (df['prob_exp'] > df['prob_noexp']))
    )

    # ذخیره
    out_path = csv_path if OVERWRITE else csv_path.with_stem(csv_path.stem + "_new")
    df.to_csv(out_path, index=False)
    print(f"✅ {csv_path.name} updated → {out_path.name}")

def main():
    # همهٔ فایل‌های csv موجود در پوشه
    for csv in sorted(EVAL_DIR.glob("*.csv")):
        update_file(csv)

if __name__ == "__main__":
    main()
```
