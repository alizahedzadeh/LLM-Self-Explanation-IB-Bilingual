```python
import json

# Load the dataset
with open('problems.json', 'r', encoding='utf-8') as file:
    dataset = json.load(file)

# Example: Accessing a specific question by ID
question_id = list(dataset.keys())[0]  # Get the first question ID
question_data = dataset[question_id]  # Access the data for that question

# Print question details
print(f"Question ID: {question_id}")
print(f"Question: {question_data['question']}")
print(f"Choices: {question_data['choices']}")
print(f"Answer: {question_data['answer']}")
print(f"Hint: {question_data['hint']}")
print(f"Grade: {question_data['grade']}")
print(f"Solution: {question_data['solution']}")
```

```python
filtered_data = {
    qid: data for qid, data in dataset.items()
    if data['split'] == 'test' and not data['image']  # Only test set and no images
}

# Print some filtered results
print(f"Total filtered questions: {len(filtered_data)}\n")
for qid, data_entry in list(filtered_data.items())[:5]:  # Display first 5 questions
    # Print each part clearly
    print("---- DATA ENTRY DETAILS ----")
    print(f"Question: {data_entry['question']}")
    print(f"Choices: {data_entry['choices']}")
    print(f"Answer (index): {data_entry['answer']} ({data_entry['choices'][data_entry['answer']]})")
    print(f"Hint: {data_entry['hint'] or 'No hint provided'}")
    print(f"Image: {data_entry['image'] or 'No image provided'}")
    print(f"Task: {data_entry['task']}")
    print(f"Grade: {data_entry['grade']}")
    print(f"Subject: {data_entry['subject']}")
    print(f"Topic: {data_entry['topic']}")
    print(f"Category: {data_entry['category']}")
    print(f"Skill: {data_entry['skill']}")
    print("Lecture:")
    print(data_entry['lecture'])
    print("Solution:")
    print(data_entry['solution'])
    print(f"Data Split: {data_entry['split']}")
    print("----------------------------")
```
