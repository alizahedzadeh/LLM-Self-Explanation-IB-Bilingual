```python
import re

def clean_prediction(pred):
    if pd.isna(pred):
        return pred
    # extract A/B/C/D if inside brackets [A]
    try:
        match = re.search(r"\[?([ABCD])\]?", str(pred).strip())
        if match:
            return match.group(1)
    except:
        return str(pred).strip()  # fallback: keep as-is

# Apply to each dataframe
for df in [df_gemini]:
    df["prediction"] = df["raw_output"].apply(clean_prediction)

# Quick check
print(df_gemini["prediction"].unique()[:10])
```

```python
import re

def extract_explanation(text: str) -> str | None:
    """
    Extract explanation text from raw_output.
    Handles both well-formed and incomplete tags.
    """
    if not isinstance(text, str):
        return None
    
    # Case 1: Properly closed </explanation>
    match = re.search(r"<explanation>\s*(.*?)\s*</explanation>", text, re.DOTALL | re.IGNORECASE)
    if match:
        return match.group(1).strip()
    
    # Case 2: Incomplete / truncated explanation tag
    match = re.search(r"<explanation>\s*(.*)", text, re.DOTALL | re.IGNORECASE)
    if match:
        return match.group(1).strip()
    
    return None

# Apply to all rows (overwrite / refill explanation)
df_gemini["explanation"] = df_gemini["raw_output"].apply(extract_explanation)
```
